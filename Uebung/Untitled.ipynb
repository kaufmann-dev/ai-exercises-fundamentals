```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris
```

```python
X = iris["data"]
y = iris["target"]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=3, random_state=12)
```

```python
X_test.shape
```

```python
clf = RandomForestClassifier(max_depth=5, n_estimators=50)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
#df = pd.DataFrame(zip(y_pred, y_test), columns=["predicted", "actual"])
df = pd.DataFrame({"predicted - y_pred":y_pred, "actual - y_test":y_test})
df.head()
```

```python
df["correct"] = df.apply(lambda row: 1 if row["predicted - y_pred"] == row["actual - y_test"] else 0, axis=1)
df.head()
```

```python
#df["correct"].count()
sum(df["correct"])/df["correct"].count()*100
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
