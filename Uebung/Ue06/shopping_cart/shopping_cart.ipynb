```python
with open('shopping_cart.txt') as f:
    read_data = f.read()
    lines = open('shopping_cart.txt').readlines()
    firstline = lines[0]
    otherlines = lines
    otherlines.pop(0)
    
    firstarray=[]
    
    for columns in firstline.split(';'):
        firstarray.append(columns)
    
    temp=[[],[],[],[]]
    tempy=0
    
    for line in otherlines:
        for columns in line.strip('\n').split(';'):
            temp[tempy].append(columns)
            tempy+=1
        tempy=0
        
    
        #otherarray.append(temp)
        #temp=[]
        #var tempx++
    
    zip_iterator = zip(firstarray, temp)
    a_dictionary = dict(zip_iterator)
    print(a_dictionary)
    
    
```

```python
with open('shopping_cart2.txt', 'w') as h:
    h.write("#Nr;Gesamtpreis;Bezeichnung\n")
        
    i=0
    while i<10:
        for key, value in a_dictionary.items():
            h.write(value[i]+";")
        h.write("\n")
        i+=1
        
```
