```python
class MissingData(Exception): #implemented
    pass
class WrongDataType(Exception): #implemented
    pass
class FileDoesNotExist(Exception): #implemented
    def __init__(self, path = "file"):
        if path == "file":
            self.message = "file does not exists"
        else:
            self.message = f"file {path} does not exists"
        super().__init__(self.message)
class MissingColumn(Exception): #implemented
    pass
class NoWriteRight(Exception): #implemented
    def __init__(self, path = "file"):
        if path == "file":
            self.message = "no writing right to file"
        else:
            self.message = f"no writing right to file {path}"
        super().__init__(self.message)
```

```python
with open('shopping_cart.txt') as f:
    lines = f.readlines()
    firstline = lines[0]
    otherlines = lines
    otherlines.pop(0)
    
    firstarray=[]
    
    for columns in firstline.split(';'):
        firstarray.append(columns.strip('\n'))
    
    temp=[[],[],[],[]]
    tempy=0
    
    for line in otherlines:
        for columns in line.strip('\n').split(';'):
            temp[tempy].append(columns)
            tempy+=1
        tempy=0
    
    zip_iterator = zip(firstarray, temp)
    a_dictionary = dict(zip_iterator)
    print(a_dictionary)
    
    
```
